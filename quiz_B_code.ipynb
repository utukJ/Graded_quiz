## Course Instructions:


To answer some questions, you will need to normalize the dataset using the MinMaxScaler after removing the following columns: [“date”, “lights”]. The target variable is “Appliances”. Use a 70-30 train-test set split with a  random state of 42 (for reproducibility). Run a multiple linear regression using the training set and evaluate your model on the test set.

```
# dropping date and lights columns

dropped_energydata = energydata.drop(columns = ["date", "lights"])  
```

```
## importing relevant packages for data modeling

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error



# normalizing dataset with MinMaxScaler
scaler = MinMaxScaler()
scaled_energydata = pd.DataFrame(scaler.fit_transform(dropped_energydata), columns = dropped_energydata.columns)

# separating features from target and splitting train and test set
target = scaled_energydata["Appliances"]
features = scaled_energydata.drop(columns = ["Appliances"])

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = .3, random_state = 42)


# running a multiple linear regression 
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)

# evaluating model on test set
predictions = lin_reg.predict(x_test)
mae = mean_absolute_error(y_test, predictions).round(2)
mse = mean_squared_error(y_test, predictions).round(2)
r2 = r2_score(y_test, predictions).round(2)
rmse = np.sqrt(mean_squared_error(y_test, predictions)).round(3)
rss = np.sum(np.square(y_test - predictions)).round(2)

print("mae: {}, mse: {}, r2_score: {}, rmse: {}, rss: {}".format(mae, mse, r2, rmse, rss))
```

```
# obtaining feature weights

weights_df = pd.DataFrame(lin_reg.coef_, features.columns)
weights_df.sort_values(by = 0)
```

```
# fitting ridge and lasso regression models

from sklearn.linear_model import Ridge, Lasso


ridge_reg = Ridge(0.4).fit(x_train, y_train)
lasso_reg = Lasso(0.001).fit(x_train, y_train)



# evaluating the models

predictions = ridge_reg.predict(x_test)
mae = mean_absolute_error(y_test, predictions).round(2)
mse = mean_squared_error(y_test, predictions).round(2)
r2 = r2_score(y_test, predictions).round(2)
rmse = np.sqrt(mean_squared_error(y_test, predictions)).round(3)
rss = np.sum(np.square(y_test - predictions)).round(2)

print("Ridge evaluation: ")
print("mae: {}, mse: {}, r2_score: {}, rmse: {}, rss: {}".format(mae, mse, r2, rmse, rss))

predictions = lasso_reg.predict(x_test)
mae = mean_absolute_error(y_test, predictions).round(2)
mse = mean_squared_error(y_test, predictions).round(2)
r2 = r2_score(y_test, predictions).round(2)
rmse = np.sqrt(mean_squared_error(y_test, predictions)).round(3)
rss = np.sum(np.square(y_test - predictions)).round(2)


print("Lasso evaluation: ")
print("mae: {}, mse: {}, r2_score: {}, rmse: {}, rss: {}".format(mae, mse, r2, rmse, rss))
```
